# correlation
The second step towards the construction of an index is to construct a correlation matrix of all the involved indicators.

```python
df4.corr(method='pearson')
```

```python
df4['Z_Assests'] = (df4['HH with Assets']-df4['HH with Assets'].mean())/df4['HH with Assets'].std()
```

```python
df4['Z_Assests']
```

```python
df4['Z_SCST'] = (df4['Percent_SC_ST _Pop']-df4['Percent_SC_ST _Pop'].mean())/df4['Percent_SC_ST _Pop'].std()
```

```python
df4['Z_SCST']
```

```python
df4['Z_ILL'] = (df4['Percent_ILL_Pop']-df4['Percent_ILL_Pop'].mean())/df4['Percent_ILL_Pop'].std()
```

```python
df4['Z_ILL']
```

```python
df4['Z_WP'] = (df4['Workforce_Participation']-df4['Workforce_Participation'].mean())/df4['Workforce_Participation'].std()
```

```python
df4['Z_WP']
```

```python
df4
```

```python
df4.describe()
```

```python
df4['SED']= - df4['Z_Assests'] + df4['Z_SCST'] + df4['Z_ILL'] - df4['Z_WP']
```

```python
df4[['Ward','Name','SED']]
```

```python
df5 = pd.read_excel("Advance Data Analysis.xlsx", sheet_name="Infrastructure_New")
```

```python
df5['Ward'] = df5["Ward"].astype('category')
```

```python
df5.corr(method='pearson')
```

```python
df5['Z_Water'] = (df5['Tap Water_Treated']-df5['Tap Water_Treated'].mean())/df5['Tap Water_Treated'].std()
```

```python
df5
```

```python
df5['Z_Walls'] = (df5['Concrete Walls']-df5['Concrete Walls'].mean())/df5['Concrete Walls'].std()
```

```python
df5
```

```python
df5['Z_Lighting'] = (df5['Electric lighting']-df5['Electric lighting'].mean())/df5['Electric lighting'].std()
```

```python
df5
```

```python
df5['Z_Drainage'] = (df5['Closed Drainage']-df5['Closed Drainage'].mean())/df5['Closed Drainage'].std()
```

```python
df5
```

```python
df5['ID']= - df5['Z_Water'] - df5['Z_Walls'] - df5['Z_Lighting'] - df5['Z_Drainage']
```
